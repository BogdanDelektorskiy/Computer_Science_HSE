## List comprehensions (списковые включения)

```python
(func(arg) for arg in iterable if expr(arg))
```

```python
a = []  # создания списка
for i in range(5):
  a.append(i ** 2)
print(a)
```

Альтернатива через списковые включения, подход похож на тернарный оператор. Мы создаем список в которм будут значения i квадрат для каждого i для range от 5. Делает тоже самое, что и код выше.

Создание списковых включений несколько более эффективно с точки зрения времени и использования памяти, нежели обычное создание списка и его пополнение.

Питон на этапе чтения строки может понять сколько в результирующем списке будет элементов и поэтому сразу выделит память под нужное количество элементов.

```python
a_copmr = [i ** 2 for i in range(5)]
print(a_copmr)
```

Усложним задачу:

```python
a = []
for i in range(10):
  if i ** 2 % 10 != 4:
    a.append(i ** 2)
print(a)

a_copmr = [i ** 2 for i in range(10)]
print(a_copmr)
```

Мы хотим добавть в список и квадрат для всех и в диапазоне до 10 если и в кв по модулю 10 отлично от 4.

```python
a_copmr = [i ** 2 for i in range(10) if i ** 2 % 10 != 4]
print(a_copmr)
```

# Сортировка в Python

- для списков:
-- sort - inplace
-- sorted - не изменяет исходный объект, возвращает отсортированную копию

- sorted принимает на вход итерируемое - iterable, кортежи, строки и тд. Возвращает всегда список
- По умолчаию сортирует по возрастанию, аргумент revers=True - по убыванию
- Не путать с reversed - возвращает обратный итератор по списку
- Сортировку можно применить к спискам, элементы которых сравнимы между собой, как правило, одного типа (можно intи float вперемешку тоже)
- Алгоритм Timsort - сортировка вставками + слиянием. O(nlogn) в среднем, O(n^2) в худшем варианте

```python
a.sort()  # метод sort  отсортирует значения внутри списка и вернет список
```

```python
t = (5, 4)  # создание кортежа
# t.sort - ошибка, корте неизменяем
# "sds".sort() - выдаст ошибку

sorted(t)
```

sorted принимает на вход любую итерируемую последовательность range, list, строка. и всегда возвращает список.

```python
t = (5, 4)
s = "sdf"
a = [i ** 2 for i in range(10) if i ** 2 % 10 != 4]

print(sorted(t), sorted(s), sorted(a), sorted(range(5, -9, -2)))
```
