## List comprehensions (списковые включения)

```python
# (func(arg) for arg in iterable if expr(arg))
```

```python
a = []  # создания списка
for i in range(5):
  a.append(i ** 2)
print(a)
```

Альтернатива через списковые включения, подход похож на тернарный оператор. Мы создаем список в которм будут значения i^2 для каждого i range от 5. Делает тоже самое, что и код выше.

Создание списковых включений несколько более эффективно с точки зрения времени и использования памяти, нежели обычное создание списка и его пополнение, так как Python на этапе чтения строки может понять, сколько в результирующем списке будет элементов и сразу выделит память под нужное количество элементов.

```python
a_copmr = [i ** 2 for i in range(5)]
print(a_copmr)
```

Усложним задачу:

```python
a = []
for i in range(10):
  if i ** 2 % 10 != 4:
    a.append(i ** 2)
print(a)

a_copmr = [i ** 2 for i in range(10)]
print(a_copmr)
```

Мы хотим добавть в список и квадрат для всех и в диапазоне до 10 если и в кв по модулю 10 отлично от 4.

```python
a_copmr = [i ** 2 for i in range(10) if i ** 2 % 10 != 4]
print(a_copmr)
```

Оба подхода создают одинаковый список, но нужно быть внимательным в плане применимости. Например, цикл ниже мы не можем написать с помощью спискового включения, но если есть возможность их использовать, то лучше использовать (встречается часто).

```python
a = []
for i in range(10):
  if i ** 2 % 10 != 4:
    a.append(i ** 2)
# else:
#   b.append(i ** 3)

print(a)
```

Как будут отсортированы эти элементы?

В лексикографическом порядке. Это естественный порядок при сортировки строк: сначала мы сравниваем первые элементы во всех строчках, если первые элементы строки a и b отличаются, то строка a меньше, чем строка b, то есть по сути для строчек где-то внутри реализован ключ, который будет для каждого символа возвращать его код.

Почему здесь используется кортеж?

При сортировки объектов значение которое будет возвращаться ключем должно быть сравнимым. Часто нам не хватает одного числа чтобы вместить всю информацию.

По умолчанию кортежи умеют сравниваться и сравниваются по элементам.



```python
# key = lambda x: foo(x)

a =["abcd", "bcd", "d", "ef"]

print(sorted(a))  # key = lambda x: tuple(ord(ch) for ch in x)

key = lambda x: tuple(ord(ch) for ch in x)
print(key(a[0]))

print(key(a[0]) < key(a[1]))  # кортежи умеют сравниваться
```

*tuple (кортеж) — это неизменяемая последовательность объектов. Это значит, что после создания кортежа вы не можете изменять его элементы, добавлять новые или удалять существующие. Кортежи могут содержать элементы разных типов, включая другие кортежи и даже списки*

```python
print(ord("h"))  # ord позволяет посмотреть код символа
```

Для кортежей все операции сравнения, = и != определены и сравнивают поэлементно. Именно поэтому, когда мы пишем пользовательские ключи для сортировки, они могут возвращать как числа, так и, зачастую, кортежи.

Итерируемое - это то, по чему мы можем пройтись, например, с помощью цикла for или созданию листа или кортежа, потому что, по сути, он делает тоже самое, что и for: проходит по элементам и не возвращает один элемент, а возвращает этот элемент и добавляет его в список или кортеж.

Теперь в исходном итерируемом объекте сначала будет положительное число, потом отрицательное, их относительный порядок сохраняется.

*Итерируемый объект — это объект, который поддерживает протокол итерации. Это означает, что вы можете перебрать его элементы с помощью циклов, таких как for, или с помощью функций, которые требуют итерируемый объект, например, list(), tuple(), set(), sum() и многие другие.*

Здесь нас интересует в первую очередь оператор **itemgetter**. Он позволяет брать по индексу какие-то значения.


> After f = itemgetter(2), the call f(r) returns r[2]

Еще есть функция **cmp_to_key**. Если мы очень хотим написать функцию, которая будет сравнивать два элемента и возвращать для них 1, -1 или 0 (как это делается в С), и мы хотим использовать вот эту функцию вместо классического ключа. Есть смысл так сделать, если какое-то очень сложное сравнение.

```python
import functools

def mycmp(a, b):
  print("comparing ", a, " and", b)
  if a > b:
    return 1
  elif a < b:
    return -1
  else:
    return 0

print(sorted([1, 2, 4, 2], key=functools.cmp_to_key(mycmp)))
```

Напишем функцию **range**.

Возникает вопрос, что мы хотим вернуть? Текущее значение в основной цикл, но при этом мы не совсем знаем, как это сделать, потому что если мы сделаем return i, то получится, что мы возратили одно число и все, и функция myRange вернула одно число.

Для того чтобы многократно возвращать какое-то значение в функцию необходимо использовать специальное ключевое слово **yield**. Это некий аналог return, котоый позволяет выходить из функции, возвращая это значение (yield i), и потом, когда вызывается следующий раз метод next, будет происходить  возврат в эту точку.

Если мы вызываем функцию несколько раз, естественно, мы будем входить в ее начало, это если мы вышли из нее с помощью return, а если мы выходим через yield, то мы возвращаемся в то место откуда выщли.

```python
def myRange(x):
  i = 0
  while i < x:
    yield i
    i += 1

for i in myRange(5):
  print(i)

print(dir(myRange))
print(dir(myRange(5)))

print(type(myRange))
print(type(myRange(5)))
```

**Любой генератор является итератором**, по скольку у него по орпделению есть какое-то внутренее состояние и метод next

Генераторные выражения

Мы могли бы ожидать, что это будет кортеж, но нет. Если так подобное выражение заключено в скобки, то это означает, что создается генератор. myRange и выражение ниже - это генераторы и принцип работы и использования у них абсолютно одинаков.

```python
a = (x for x in range(5) if x % 2 == 1)

for i in a:
  print(i)
```

Каждый проход **for** будет брать следующий элемент. Мы можем сделать и вручную - сделать next от a, поскольку у него есть next.

Результат работы можем сохранить, например, спичок, с помощью list, как например, ниже, мы можем приземлить list, то есть список у нас сам пробежится. Вычитает все значения, поскольку это у нас итератор, и при создании листа будет проходить все эти шаги итерации. Он вычитает 0, 1, 2, 3, 4 и записывает в список, дальше мы можем с этим списком работать.

Но обычно если есть генератор, то это означает, что мы не хотим записывать все эти значения в память, мы наоборот, хотим генерировать их на лету.



```python
def myRange(x):
  i = 0
  while i < x:
    yield i
    i += 1

print(list(myRange(5)))
```

Если бы написали это в [ ], то это было бы создано в памяти как список, в данном случае мы, очевидно, не хотим этого делать. В этом случае у нас будет генератор, который хранит внутри себя текущее значение х, и он будет возвращать какое-то значение.

Для того чтобы приземлить это в память мы можем точно также сделать list.

Представим себе, что мы хотим написать генератор, который хочет нам генерировать числа заданной длины, цифры не убывают и старшая цифра не превосходит какого-то параметра, например, если мы хотим сделать множество чисел у которых длина 2 и максимальная цифра 3. Как мы можем это сделать?

```python
def genDecDigs (cntDigits, maxDigit):
  if cntDigits > 0:  # пока количесвто цифр > 0
    for nowDigit in range(maxDigit + 1):  # перебираем текущую цифру
      for tail in genDecDigs(cntDigits - 1, nowDigit):  # выбираем хвост для этой цифры
        yield nowDigit * 10**(cntDigits - 1) + tail  # составляем новое число
  else:
    yield 0
print(*genDecDigs(2, 3))
```

Иногда вот такого вида генераторные функции называются **сопрограммами**, поскольку это независимая, даже не подпрограмма, а сопрограмма, которая выполняется во многом независимо от нашей, поскольку у нее есть свое внутренее состояние, и которая обменивается с нашей программой информацией.

```python
def myRange(x):
  i = 0
  while i < x:
    yield i == i + 1
```
