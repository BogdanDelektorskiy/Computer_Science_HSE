## Декораторы



- gререквизиты:
 - функции в  Python являются объектами 1 класса, соответственно, их можно возвращать из другой функции или передавать в качестве аргумента в функцию
  - в Python можно определять вложенные функции
- декораторы - это, по сути, "обертка", которая жает нам воззможность изменить поведение функции, не изменяя ее код
- один и тот же декоратор можно применить к разным функциям
- к одной функции можно применить несколько декораторов последовательно

**Объект первого класса**

Объект называют "объектом первого класса", если он:

- может быть сохранен в переменной или структурах данных
- может быть передан в функцию как аргумент
- может быть возвращен из функции как результат
- може быть создан во время выполнения программы
- внутренне самоопознаваем (независим от именования)


В нашем случае функция как раз таки является объектом первого класса. Это важно, так как почти во всех других языках функция не является объектом первого класса.

```python
def decorator(function_to_decorate):
  def the_wrapper_around_the_original_function():
    print("Before")
    res = function_to_decorate()
    print("After")
    return res
  return the_wrapper_around_the_original_function

@decorator
def our_function():
  print("Hello")

# our_function_decorated = decorator(our_function)

our_function()
```

Использование декоратора через @ переопределяет вызов функции, теперь наша функция это декоратор от нашей функции.

На этапе вызова функции мы функцию декоратор даже не затронули, то есть самувнешнюю функцию.

## Парадигмы программирования

- совокупность идей и понятий, определяющих стиль написания программ
- большинство современных языков мультипарадигменные (в т.ч. и Python)
- исмперативное (структурное, процедурное) - программы есть последовательность инструкций, которые читают и записывают данные из памяти
- декларированное - описывает желаемый результат, а не способ его получения (SQL)
- объектно-ориентированное - программы используют объекты, при этом у объекта есть состояние и методы для изменнения этого состояния
- функциональное - задача разбивается на набор функций. В идеале, функции только принимают параметры и возвращают значения, не изменяя состояния объектов и программы (huskel)
- логическое программирование

### Функциональное программирование:

- + лгеко проводить декомпозицию, откладку и тестирование программ
- + удобно распараллеливать вычесления (векторизация, контейенизация)
- векторизация - одни и те же действия выполняются над большим набором данныхю Части данных можно раскидать по разным вычислительным устройствам, а затем собирать результаты в один объект
- конвейеризация - разбиение вычислений на несколько этапов, причем данные поступают на следующий этап обработки по времени готовности

```python
l = range(7)

print(list(map(lambda x: x**2, filter(lambda x: x % 2 == 1, l))))
```

```python
l = range(7)

print(list(zip(map(lambda x: x**2, filter(lambda x: x % 2 == 1, l)), map(lambda x: x + 100, range(15, 20)))))
```

```python
from functools import reduce

print(reduce(lambda x, y: x + y, [5, 6, 7, 8]))
```

```python
from functools import reduce

print(reduce(lambda x, y: x + y, ["5", "6", "7", "8"]))
```

```python
def gcd(a, b):
  while b != a:
    a, b = b, a % b
  return a

print(reduce(gcd, map(int, input().split())))
```

```python
from itertools import accumulate

def gcd(a, b):
  while b != a:
    a, b = b, a % b
  return a

print(list(accumulate(map(int, "324 534 24 12 72".split()), gcd)))
```
