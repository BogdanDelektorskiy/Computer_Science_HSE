## А как с этим работает for?

- внутренний механизм цикла for сначала вызывает метод __iter() объекта. Так что, если передан итерируемый объект, создается итератор. После этого применяетс метод __next() до тех пор, пока не будет возбуждено исключение StopIteration
- поскольку метод __iter() итератора возвращает сам итератор, то после перебора циклом for объект исчерпывается. То есть получить данные из итератора можно только один раз
- в случае с коллекциями это не так. Здесь создается другой объект - итератор. Он, а не итерируемый объект, отдается на обработку циклу for

## Генераторы

- a = (x for x in range(5)if x % 2 == 1) - генераторное выражение
- бесконечный генератор: while True, внутри которого yield - генераторная функция
- метод next - берет следующее значение из генератора
- throw() - бросить исключение внутри генератора. Исключение будет возбужденно из того выражение yield, где генератор последний раз остановился
- close() - закртыь генератор. Бросает внутри генератора особое исключение GeneratorExit. после этого next() будет возвращать StopIteration

```python
def gen():
  while True:
    yield 1

g = gen()
for i in g:
  print(i)
```

Бесконечный цикл

Есть еще возможность, которая позволяет управлять состоянием генератора из вне: throw() и close()

```python
def gen():
  i = 1
  while True:
    yield i
    i += 1

g = gen()
for i in g:
  print(i)
  if i == 8:
    g.throw(Exception())
```

Мы получаем исключение, но если мы пишем так, то мы говорим про то что это может быть любое исключение, в том числе системное. И вот мы любое исключение поймаем если напишем BaseException. Но как правило так не рекомендуетя делать, потому что если, например, приходит Cntr+C в какой-то момент, то наверное мы должны завершиться. При том что, на самом деле у нас бы возникла проблема в индексации.

Чтобы не игнорировать при обработке все системные исключения мы пишем exept и Exception as e, здесь не будет никаких дополнительных деталей, мы просто видем какое-то исключение которое мы можем обрабатывать.

попробуем обрабатывать исключения последовательно:

```python
try:
  1 / 0
  a[5]
except Exception as e:
  print(e, type(e))
except ZeroDivisionException as e:
  print(e)
```

Сначала пишем более узкие группы исключений, и лишь в цонце пишем более обще обработчик исключений для тех случаев, если мы не ожидали этого.

Если мы хотим обрабатывать какие-то исключения одинавково, мы их заключаем в ( ) и разделяем запятой.

```python
try:
  1 / 0
  a[5]
except Exception as e:
  print(e, type(e))
except ZeroDivisionException as e:
  print(e)
else:
  print("else")  # если нет исключений
finally:
  print("finlly")  # обработка, не зависимо от наличия исключений
    # finally, например, закрываются открыте файлы
```

## Передача значений в генератор
- передача значений в генератор -send- отправляем значения на обработку
- первый раз нельзя посылать в генератор данные, пока мы не прокрутили его до первого yield. Нужно либо вызыватьnext(g) или g.send(None) - это одно и тоже.







```python
def generator(x):
  while True:
    x = yield x + 1  # возвращает значенние, которое отправляет send

g = generator(5)
print(g.send(None))  # nexd(g) == g.send(None)
print(g.send(10))
print(g.send(100))
```

## Модуль typing

- расширение стандартных типов, используется [ ]
- Optional[type]: указанный тип или None
- Any - любой тип (в отличие от object, который определяет множество методов)
- Union - использование типов из множества, Optional[T] эквивалентна Union[T, None]
- коллекции: иногда недостаточно указывать list, tuple и прочее, а надо конкретизировать тип элементов в клллекции

```python
def foo(x: int) -> [int|None]:
  if x >= 0:
    return x
# неявно здесь есть return None
```

```python
from __future__ import annotations
from typing import Optional

def foo(x: int) -> Optional[int]:
  if x >= 0:
    return x
  return None
```

значение возвращается опциоально, то есть оно либо возвращается либонет (то есть возвращает None)

```python
from __future__ import annotations
from typing import Any

def foo(x: int) -> Any:
  if x >= 0:
    return x
  return None
```
