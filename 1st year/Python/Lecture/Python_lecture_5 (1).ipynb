## List comprehensions (списковые включения)

```python
# (func(arg) for arg in iterable if expr(arg))
```

```python
a = []  # создания списка
for i in range(5):
  a.append(i ** 2)
print(a)
```

Альтернатива через списковые включения, подход похож на тернарный оператор. Мы создаем список в которм будут значения i квадрат для каждого i для range от 5. Делает тоже самое, что и код выше.

Создание списковых включений несколько более эффективно с точки зрения времени и использования памяти, нежели обычное создание списка и его пополнение.

Питон на этапе чтения строки может понять сколько в результирующем списке будет элементов и поэтому сразу выделит память под нужное количество элементов.

```python
a_copmr = [i ** 2 for i in range(5)]
print(a_copmr)
```

Усложним задачу:

```python
a = []
for i in range(10):
  if i ** 2 % 10 != 4:
    a.append(i ** 2)
print(a)

a_copmr = [i ** 2 for i in range(10)]
print(a_copmr)
```

Мы хотим добавть в список и квадрат для всех и в диапазоне до 10 если и в кв по модулю 10 отлично от 4.

```python
a_copmr = [i ** 2 for i in range(10) if i ** 2 % 10 != 4]
print(a_copmr)
```

Оба подхода создают одинаковый список, но нужно быть внимательным в плане применимости. Например, цикл ниже мы не можем написать с помощью спискового включения. Но если есть возможность их использовать, то лучше использовать (встречается часто).

```python
a = []
for i in range(10):
  if i ** 2 % 10 != 4:
    a.append(i ** 2)
# else:
#   b.append(i ** 3)

print(a)
```

Как будут отсортированы эти элементы?
В лексикографическом порядке, это естественный порядок при сортировки строк. Сначала мы сравниваем первые элементы во всех строчках, если первые элементы строки a и b отличаются, то строка a меньше, чем строка b.
То есть по сути для строчек где-то внутри реализован ключ, который будет для каждого символа возвращать его код.

Почему здесь используется кортеж?
При сортировки объектов значение которое будет возвращаться ключем должно быть сравнимым. Часто нам не хватает одного числа чтобы вместить всю информацию.

По умолчанию кортежи умеют сравниваться и сравниваются по элементам.



```python
# key = lambda x: foo(x)

a =["abcd", "bcd", "d", "ef"]

print(sorted(a))  # key = lambda x: tuple(ord(ch) for ch in x)

key = lambda x: tuple(ord(ch) for ch in x)
print(key(a[0]))

print(key(a[0]) < key(a[1]))  # кортежи умеют сравниваться
```

tuple (кортеж) — это неизменяемая последовательность объектов. Это значит, что после создания кортежа вы не можете изменять его элементы, добавлять новые или удалять существующие. Кортежи могут содержать элементы разных типов, включая другие кортежи и даже списки

```python
print(ord("h"))  # ord позволяет посмотреть код символа
```

Для кортежей все операции сравнения и равное и != определены и сравнивают поэлементно. Именно поэтому, когда мы пишем пользовательские ключи для сортировки они могут возвращать как числа так и зачастую кортежи.

У нас есть генераторное выражение которое будет возвращать все значения, потом все эти значения будут складываться в некий внутренний объект и уже внутри они будут сортироваться.

Итерируемое - это то по чему мы можем пройтись, например, с помощью цикла for или созданию листа или кортежа, потому что по сути он делает тоже самое что и for: проходит по элементам и не возвращает один элемент, а возвращает этот элемент и добавляет его в список или кортеж.

Как бы мы могли сделать то же самое нео не модифицировать а отсортировать по квадрату

```python
print(sorted(range(-5, 6), key=lambda x: x **2))
```

Отсортирован именно так потомму что значение квадрата для -х и для х будет одинаковым и при этом в рамках  нашего range сначала всегда встречается отрицатеьное число потмо положительное, поэтому их относительный порядок сохранится в сиду устойчивости.

Теперь в исходном итеребл объекте сначала будет положительное число, потом отрицательное, их относительный порядок сохраняется.

Мы всегда можем поменять этот порядок, и сказать что если х квадрат одинаковый, то сортируй по х

```python
print(sorted(range(5, -6, -1), key=lambda x: x **2))
```

здесь нас интересует в первую очередь оператор **itemgetter**. Он позволяет брать по индексу какие-то значения.

After f = itemgetter(2), the call f(r) returns r[2].

After g = itemgetter(2, 5, 3), the call g(r) returns (r[2], r[5], r[3])

```python
a = [[3, 4], [5, 7], [7, 2], [1, 4]]
print(sorted(a, key=operator.itemgetter(1)))
```

Что это делает? берет по заданным индексам элементы и складывает их в кортеж. Заявляет что это более эффективно key=lambda x: x[1]

Еще есть функция cmp_to_key. Если мы очень хотим написать функцию которая будет сравнивать два элемента и будет возвращать для них 1, -1 или 0 (как это делается в С) и мы хотим использовать вот эту функцию вместо классического ключа. если какое-то очень сложное сравнение, то возмоожно есть смысл так сделать

```python
import functools

def mycmp(a, b):
  print("comparing ", a, " and", b)
  if a > b:
    return 1
  elif a < b:
    return -1
  else:
    return 0

print(sorted([1, 2, 4, 2], key=functools.cmp_to_key(mycmp)))
```
