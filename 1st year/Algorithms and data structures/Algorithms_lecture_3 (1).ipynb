*расчет факториала числа n:*

*если n равно 0 или 1, функция возвращает 1 (база рекурсии). В противном случае, функция возвращает произведение n на факториал числа n-1*

```python
def f(n):
  if n in [0, 1]:
    return 1
  else:
    return n *f(n -1)

print(f(5))
```

## Генерация перестановок

abc - какая-то перестановка

abc, acb, bac, bca, cab, cba - как их сгенерировать? Рекурсия

image.png

```python
def gena(head: str, tail: str) -> None:
  if len(head) == 0:
    print(tail)
  else:
    for i in range(len(head)):
      gena(head[:i] + head[i +1:], tail + head[i])
gena("abc","")
```

```python
s = "abc"
n = len(s)
cnt = [0] * n
ans = []

# получить все перестановки используя каждую букву 2 раза
def gena():
  global cnt, ans, n, s
  if len(ans) == n:
    print(''.join(ans))
  else:
    for i in range(n):
      if cnt[i] < 2:
        ans.append(s[i])
        cnt[i] +=1
        gena()
        ans.pop()
        cnt[i] -= 1
gena()
```

*Операция .pop():*
**ans.pop():**
*Удаляет и возвращает последний элемент из списка ans.*
*Это действие происходит после того, как функция gena() завершает один из рекурсивных вызовов, что означает возврат к предыдущему состоянию.*
*Это необходимо для отката изменений, когда происходит возвращение из рекурсии, чтобы проверить другие комбинации символов.*
*Синтаксис pop():*
*Метод pop() вызывается на объекте (в данном случае — это список ans):*
*last_element = ans.pop()*
*Это приведет к тому, что last_element станет равным последнему элементу списка, а список ans потеряет этот элемент.*
