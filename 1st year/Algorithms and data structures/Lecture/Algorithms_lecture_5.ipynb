## Поиск в массиве (целочисленный)

Есть массив целых чисел отсортированный пл неубыванию.

Запрос вида: найти первое вхождение элемента -key. При этом размер массива размер массива n*10^6. Количество запросов q примерно 10^5.

Если мы будем просто линейным поиском идти и смотреть первое вхождение. Это не будет адекватно работать, так как асимптотика будет в худшем случае 10^11 - это нас вообще никак не устраивает.

Используется бинарный поиск (двоичный поиск, дихотономический поиск).

Создадим два каких-то фиктивных элемента (их реально нет и не будет). Указатель l поставим на (-1), а указатель r = a.size()

Ищем ключ 6 key=6

```python
arr = [2, 2, 2, 4, 4, 4, 4, 6, 7, 8, 8, 9]


def bin_first(arr, key):
  L = -1
  R = len(arr)
  while L + 1 < R:
    m = (L+ R) // 2
    if arr[m] < key:
      L = m
    else:
      R = m
  if R != len(arr) and arr[R] == key:
    return R
  else:
    return -1

print(bin_first(arr, 2))
print(bin_first(arr, 4))
print(bin_first(arr, 9))
print(bin_first(arr, 5))




```

## Вещественный бинарный поиск

Предстваим у нас есть какая-то функция f(x) и мы знаем как она выглядит, она четко определена на отрезке [a, b]. Мы хотим найти корень этой функции. Как его найти?

image.png

Сужаемся к этой точке до тех пор пока условный eps = b - a не станет равен примерно 10^-6.

Работает если:
- только один корень на [a, b]
- функция определена на [a, b]

Если на [a, b ] два и более корней - отделить их.

```python
def f(x:float) -> float:
  return x * x - 3

left = 0
right = 3
cnt = 0
while right - left > 0.0000001:
  cnt += 1
  m = (left + right) / 2
  if f(left) * f(m) > 0:
    left = m
  else:
    right = m

x = (left + right) / 2
print(cnt)
print(x)

```

Итоговая асимптотика - O(log (b - a) / 2)
